```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Negaci√≥n

```python
df = pd.read_excel("Datos_Neg.xlsx", engine="openpyxl")
df.columns = df.columns.str.strip()  

print(df)
```

```python
plt.figure(figsize=(8, 5))

x_pos = np.arange(len(df))

bar_width = 0.25 

colors = ['blue', 'green', 'red']

plt.bar(x_pos - bar_width, df["Train F1 Score"], width=bar_width, color=colors[0], alpha=0.7, label="Train F1")
plt.bar(x_pos, df["Jabberwockie F1 Score"], width=bar_width, color=colors[1], alpha=0.7, label="Jabberwockie F1")
plt.bar(x_pos + bar_width, df["Consistency F1 Score"], width=bar_width, color=colors[2], alpha=0.7, label="Consistency F1")

plt.xticks(x_pos, df["Cantidad palabras por clase"])  

plt.xlabel("Cantidad de palabras por clase")
plt.ylabel("F1 Score")
plt.title("Comparaci√≥n de F1 Scores por Cantidad de Palabras")
plt.ylim(0, 1)  
plt.legend()  
plt.grid(axis="y", linestyle="--", alpha=0.5)  

plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4), tight_layout=True)

metrics = ["Train F1 Score", "Jabberwockie F1 Score", "Consistency F1 Score"]
colors = ['blue', 'green', 'red']
titles = ["Train F1 Score", "Jabberwockie F1 Score", "Consistency F1 Score"]

for i, ax in enumerate(axes):
    sns.barplot(x=df["Cantidad palabras por clase"], y=df[metrics[i]], color=colors[i], ax=ax)
    ax.set_title(titles[i])
    ax.set_xlabel("Cantidad de palabras por clase")
    ax.set_ylabel("F1 Score")
    ax.set_ylim(0, 1) 
    ax.grid(axis="y", linestyle="--", alpha=0.5) 

plt.show()
```
